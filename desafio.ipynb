### Desafio 1 

1) Trazer a média das vendas.
2) Criar um gráfico de barras vertical mostrando o mês de referência e o valor.
3) Criar um gráfico de linhas mostrando o mês de referência e o valor.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
# Etapa 1 - Trazer a média das vendas.

print(f"O valor médio do faturamento foi de R$ {df_faturamento.valor.mean():.2f}")
```

```python
# Etapa 2 - Criar um gráfico de barras vertical mostrando o mês de referência e o valor.

# 1) Conferir os dados

df_faturamento.info()
```

```python
# Transformação do objeto para Objeto de Data
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
# Formatando para pegar apenas o mês com .dt.month
df_faturamento['data_ref'] = df_faturamento['data_ref'].dt.month
```

```python
# Gráfico Final 
# Legendas: 1e6 no canto superior esquerdo diz que está em notação ciêntífica na casa do milhão 10^6
df_faturamento.plot.bar(x='data_ref', y='valor')
```

```python
# Etapa 3 - Criar um gráfico de linhas mostrando o mês de referência e o valor.

# só precisou disso pq a lógica já foi montada nos anteriores, tbm está em notação para economizar espaço.
df_faturamento.plot.line(x='data_ref', y='valor')
```
